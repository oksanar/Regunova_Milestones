<h3>Graph where bars are stacked</h3>

```python
insta = {'Days': [2, 3, 4, 5], 
              'From Home': [2586, 2727, 2085, 2700], 
              'From Hashtags': [1028, 1838, 1188, 621], 
              'From Explore': [619, 1174, 0, 932]}

df = pd.DataFrame(insta)
```

```python
# Set the width of the bars
barWidth = 0.11
x = np.arange(4)

# Set the position of the bars on the x-axis
r1 = np.arange(len(df['Days']))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

# Create the bar plots
plt.bar(r1, df['From Home'], color='red', width=barWidth, edgecolor='white', label='From Home')
plt.bar(r2, df['From Hashtags'], color='yellow', width=barWidth, edgecolor='white', label='From Hashtags')
plt.bar(r3, df['From Explore'], color='pink', width=barWidth, edgecolor='white', label='From Explore')

# Add x-axis and y-axis labels and a title
plt.xlabel('Days')
plt.ylabel('Impressions')
plt.title('Instagram')

# Add legend
plt.legend()

# Show the chart
plt.show()
```
